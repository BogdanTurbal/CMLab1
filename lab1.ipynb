```python
import matplotlib.pyplot as plt
import numpy as np
import math
```

```python
def func(x):
    return x**4 - 5.74 * x**3 + 8.18 * x - 3.48
```

```python
def func_d(x):
    return 4 * x**3 - 5.74 * 3 * x**2 + 8.18
```

```python
def func_dd(x):
    return 12 * x**2 - 5.74 * 6 * x
```

```python
def sign(x):
    if x == 0:
        return 0
    return x / abs(x)
```

```python
def func_fi(x):
    return (abs((x**4 + 8.18 * x - 3.48) / 5.74) ** (1./3.)) * sign((x**4 + 8.18 * x - 3.48))
```

```python
def func_newton(x):
    return x - func(x) / func_d(x)
```

```python
def func_fi_d(x):
    return 1/3 * 1 / (pow(pow((x**4+8.18*x-3.48)/5.74, 2), 1/3)) * ((4*x**3+8.18)/5.74)
```

```python
def plot_func(func=func, a=-1.5, b=0.1):
    x = np.arange(a, b, 0.05)
    y = np.array(list(map(lambda x_: func(x_), x)))
    plt.plot(x, y)
    plt.plot([a-0.5, b], [0, 0])
    plt.show()
```

```python
def dichotomy(a, b, eps):
    i = 0
    x = -100
    print(f"initial {i}: x: {x} f: {func(x)}")
    while True:
        x_ = x
        i += 1
        x = (a + b)/2
        print(f"{i}: x: {x} f: {func(x)}")
        if sign(func(a)) == sign(func(x)):
            a = x
        if sign(func(b)) == sign(func(x)):
            b = x
        #i += 1
        if abs(x_ - x) <= eps * 2:
            break
    #print(f"{i}: x: {x} f: {func(x)}")
    return x
```

```python
print(func(-1.5), func(-1.1))
```

```python
x = dichotomy(-1.5, -1.1, eps)
print(x, func(x))
```

```python
def newton(x, eps):
    i = 0
    x = x
    print(f"initial {i}: x: {x} f: {func(x)}")
    while True:
        x_ = x
        i += 1
        x = func_newton(x)
        print(f"{i}: x: {x} f: {func(x)}")
        
        if abs(x_ - func_newton(x)) <= eps:
            break
    return x
```

```python
x = newton(-1.3, eps)
print(x, func(x))
```
